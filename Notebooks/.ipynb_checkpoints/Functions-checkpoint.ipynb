```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pylab

plt.style.use('ggplot')
pylab.rcParams['figure.figsize'] = (10, 7)
```

```python

def plot_function(x_range, f):
    y = np.vectorize(f)(x_range)
    plt.plot(x_range, y, linewidth=1.0)
    plt.xlabel('μ'.decode('utf8'))
    plt.ylabel('Wait time (seconds)')
    plt.title('Wait time by μ'.decode('utf8'))
    plt.show()
```

```python
LAMBDA = 388
def waiting_time(mu=80):
    try:
        return 1 / float(mu - LAMBDA)
    except:
        return 0

plot_function(np.arange(388.01, 400, .1), waiting_time)
```
